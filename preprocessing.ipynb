```python
#categorical =x.select_dtypes(include="object").columns
categorical=[col for col in df.columns if df[col].dtype=="object"]
categorical
```

```python
#chi2-square test
related_columns={}
not_related_columns={}

for col in categorical:
    table=pd.crosstab(df[col],y["Result"])
    chi2,p,dof,expected=chi2_contingency(table)
    if p>=0.05:
        not_related_columns[col]=p
    else:
        related_columns[col]=p

print("related_columns",related_columns)
print("\n not_related_columns",not_related_columns)
```

```python
#annova_test
numeric_col=[col for col in df.columns if df[col].dtypes!="object"]
numeric_col
```

```python
#annova_test
related_numeric_columns={}
not_related_numeric_columns={}
for col in numeric_col:
    Away=df[y["Result"]==0][col]
    Draw=df[y["Result"]==1][col]
    Home=df[y["Result"]==2][col]
    f_stats,p_value=f_oneway(Away,Draw,Home)
    if p_value>=0.05:
        not_related_numeric_columns[col]=p_value
    else:
        related_numeric_columns[col]=p_value

print("related_numeric_columns : ",related_numeric_columns)
print("not_related_numeric_columns : ",not_related_numeric_columns)
```

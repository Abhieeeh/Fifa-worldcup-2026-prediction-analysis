```python
#df["Match Date"].dtype
df["Match Date"]=pd.to_datetime(df['Match Date'],format="mixed")
df["match_day"]=df["Match Date"].dt.day
df["match_month"]=df["Match Date"].dt.month
df["match_year"]=df["Match Date"].dt.year
df.drop("Match Date",axis=1,inplace=True)
df.head()
```

```python
df["Match Time"]=pd.to_datetime(df["Match Time"],format="mixed")
df["start_time"]=(df["Match Time"].dt.hour+(df["Match Time"].dt.minute)/60)
df.drop("Match Time",axis=1,inplace=True)
```

```python
df.drop(columns=["Key Id", "Tournament Id", "tournament Name", "Match Id", "Match Name"],inplace=True)
df
```

```python
x=df.drop(columns=['Home Team Id','Home Team Code', 'Away Team Id','Away Team Code', 'Score', 'Score Penalties','Result', 'Score', 'Home Team Score', 'Away Team Score','Home Team Score Margin', 'Away Team Score Margin', 'Extra Time','Penalty Shootout', 'Score Penalties', 'Home Team Score Penalties','Away Team Score Penalties', 'Result', 'Home Team Win', 'Away Team Win','Draw','Stadium Id'])
y=df[["Result"]]
x.shape
```

```python
x.columns
x
```

```python
#categorical =x.select_dtypes(include="object").columns
categorical=[col for col in x.columns if x[col].dtype=="object"]
categorical
```

```python
#chi2-square test
related_columns={}
not_related_columns={}
for col in categorical:
    table=pd.crosstab(x[col],y["Result"])
    chi2,p,dof,expected=chi2_contingency(table)
    if p>=0.05:
        not_related_columns[col]=p
    else:
        related_columns[col]=p

print("related_columns",related_columns)
print("\n not_related_columns",not_related_columns)
```

```python
#annova_test
numeric_col=[col for col in x.columns if x[col].dtypes!="object"]
numeric_col
```

```python
#annova_test
related_numeric_columns={}
not_related_numeric_columns={}
for col in numeric_col:
    Away=x[y["Result"]==0][col]
    Draw=x[y["Result"]==1][col]
    Home=x[y["Result"]==2][col]
    f_stats,p_value=f_oneway(Away,Draw,Home)
    if p_value>=0.05:
        not_related_numeric_columns[col]=p_value
    else:
        related_numeric_columns[col]=p_value

print("related_numeric_columns : ",related_numeric_columns)
print("not_related_numeric_columns : ",not_related_numeric_columns)
```

```python
x.drop(["Stadium Name"],axis=1,inplace=True)
x.head()
```

```python
label_encoder = LabelEncoder()
```

```python
columns_to_encode = [col for col in x.columns if x[col].dtype == "object"]
for col in columns_to_encode:
    x[col] = label_encoder.fit_transform(x[col])
y["Result"]=label_encoder.fit_transform(y["Result"])
x.head()
```

```python
model = LogisticRegression()
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42, test_size=0.2)
model
```

```python
model.fit(x_train,y_train)
model
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
